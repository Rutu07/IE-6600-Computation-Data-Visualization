```python
import pandas as pd
import numpy as np
from functools import reduce
```

```python
df=pd.read_csv(r'C:\Users\rutuj\OneDrive\Documents\Northeastern University MS in Data Analytics Engineering\5. Computation & Visualization\Hackathon\COVID_19_Nursing_Home_Data_03_20_2022(3)\COVID_19_Nursing_Home_Data_03_20_2022(3)\COVID_19_Nursing_Home_Data_03_20_2022.csv')
```

```python
df.columns
```

```python
df1=df[['Week Ending', 'Federal Provider Number', 'Provider Name',
       'Provider Address', 'Provider City', 'Provider State',
       'Provider Zip Code', 'Provider Phone Number', 'County',
        'Shortage of Nursing Staff','Shortage of Clinical Staff',
        'Shortage of Other Staff','Shortage of Aides'
       ]]

df1.dropna(inplace=True)
```

```python
df1.head()
```

```python
df3=df[['Federal Provider Number','Provider Name','Provider State']]
df3.head()
df4=df3.drop_duplicates()
df4.head()
```

```python
state_wise_nursing_home_count=df4['Provider State'].value_counts().reset_index().rename(columns={'index':'Provider State','Provider State':'Count_of Nursin Home'})
state_wise_nursing_home_count.head()
```

```python
shortage_of_nursing_staff=df1.groupby(['Week Ending','Provider State','Shortage of Nursing Staff']).size().reset_index().rename(columns={0:'Count of Nursing Staff Shortage'})
shortage_of_nursing_staff=shortage_of_nursing_staff[shortage_of_nursing_staff['Shortage of Nursing Staff']=='Y']
shortage_of_nursing_staff.head()
#shortage_of_nursing_staff.shape
```

```python
shortage_of_clinical_staff=df1.groupby(['Week Ending','Provider State','Shortage of Clinical Staff']).size().reset_index().rename(columns={0:'Count of Clinical Staff Shortage'})
shortage_of_clinical_staff=shortage_of_clinical_staff[shortage_of_clinical_staff['Shortage of Clinical Staff']=='Y']
shortage_of_clinical_staff.head()
```

```python
shortage_of_other_staff=df1.groupby(['Week Ending','Provider State','Shortage of Other Staff']).size().reset_index().rename(columns={0:'Count of other Staff Shortage'})
shortage_of_other_staff=shortage_of_other_staff[shortage_of_other_staff['Shortage of Other Staff']=='Y']
shortage_of_other_staff.head()
```

```python
shortage_of_aides_staff=df1.groupby(['Week Ending','Provider State','Shortage of Aides']).size().reset_index().rename(columns={0:'Count of Aides Shortage'})
shortage_of_aides_staff=shortage_of_aides_staff[shortage_of_aides_staff['Shortage of Aides']=='Y']
shortage_of_aides_staff.head()
```

```python
dfs=[shortage_of_nursing_staff,shortage_of_other_staff,shortage_of_aides_staff,shortage_of_clinical_staff]
dataset_Q1=reduce(lambda left,right:pd.merge(left,right,on=['Week Ending','Provider State'],how='inner'),dfs)
```

```python
dataset_Q1.head()
```

```python
dataset_shortage=pd.merge(dataset_Q1,state_wise_nursing_home_count,on='Provider State',how='inner')
```

```python
dataset_shortage.head()
```

```python
dataset_shortage['Shortage of Nursing Staff(%)']=(dataset_shortage['Count of Nursing Staff Shortage']/dataset_shortage['Count_of Nursin Home'])*100
dataset_shortage['Shortage of other Staff(%)']=(dataset_shortage['Count of other Staff Shortage']/dataset_shortage['Count_of Nursin Home'])*100
dataset_shortage['Shortage of Clinical Staff(%)']=(dataset_shortage['Count of Clinical Staff Shortage']/dataset_shortage['Count_of Nursin Home'])*100
dataset_shortage['Shortage of Aides (%)']=(dataset_shortage['Count of Aides Shortage']/dataset_shortage['Count_of Nursin Home'])*100
```
