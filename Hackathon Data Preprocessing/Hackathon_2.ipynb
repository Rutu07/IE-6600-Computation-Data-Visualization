```python
import pandas as pd
import numpy as np
from functools import reduce
```

```python
df=pd.read_csv(r'C:\Users\rutuj\OneDrive\Documents\Northeastern University MS in Data Analytics Engineering\5. Computation & Visualization\Hackathon\COVID_19_Nursing_Home_Data_03_20_2022(3)\COVID_19_Nursing_Home_Data_03_20_2022(3)\COVID_19_Nursing_Home_Data_03_20_2022.csv')
```

```python
df.head()
```

```python
df['year_month']=pd.to_datetime(df['Week Ending']).dt.to_period('M')
df['month']=pd.Series(pd.DatetimeIndex(df['Week Ending']).month)
df['year']=pd.Series(pd.DatetimeIndex(df['Week Ending']).year)
```

```python
df1=df[['Week Ending', 'Federal Provider Number', 'Provider Name',
       'Provider Address', 'Provider City', 'Provider State',
       'Provider Zip Code', 'Provider Phone Number', 'County','Eye Protection No Longer Available in 7 Days', 
       'Gloves No Longer Available in 7 Days','Alcohol-Based Hand Rub (ABHR) Available',
       'Alcohol-Based Hand Rub (ABHR) No Longer Available in 7 Days',
       'N95 Respirator No Longer Available in 7 Days','Face Masks No Longer Available in 7 Days']]
df1.dropna(inplace=True)
```

```python
#Alcohol-Based Hand Rub (ABHR) Available
alcohol_sanitizer_availability=df1.groupby(['Week Ending','Provider State','Alcohol-Based Hand Rub (ABHR) Available']).size().reset_index().rename(columns={0:'Count of Eye Protection'})
alcohol_sanitizer_availability=alcohol_sanitizer_availability[alcohol_sanitizer_availability['Alcohol-Based Hand Rub (ABHR) Available']=='Y']
alcohol_sanitizer_availability.head()
```

```python
## Eye Protection Availability
```

```python
eye_protection_avail=df1.groupby(['Week Ending','Provider State','Eye Protection No Longer Available in 7 Days']).size().reset_index().rename(columns={0:'Count of Eye Protection'})
eye_protection_avail=eye_protection_avail[eye_protection_avail['Eye Protection No Longer Available in 7 Days']=='N']
eye_protection_avail.head()
#eye_protection_avail.shape
```

```python
##Gloves Availabolity
```

```python
gloves_avail=df1.groupby(['Week Ending','Provider State','Gloves No Longer Available in 7 Days']).size().reset_index().rename(columns={0:'Count of Nursing Homes having Gloves'})
gloves_avail=gloves_avail[gloves_avail['Gloves No Longer Available in 7 Days']=='N']
gloves_avail.shape
gloves_avail.head()
```

```python
## Sanitizer Availability
```

```python
#Alcohol-Based Hand Rub (ABHR) No Longer Available in 7 Days
alcohol_avail=df1.groupby(['Week Ending','Provider State','Alcohol-Based Hand Rub (ABHR) No Longer Available in 7 Days']).size().reset_index().rename(columns={0:'Count of Nursing Homes Having Sanitizer'})
alcohol_avail=alcohol_avail[alcohol_avail['Alcohol-Based Hand Rub (ABHR) No Longer Available in 7 Days']=='N']
alcohol_avail.shape
alcohol_avail.head()
```

```python
## N95 Availability
```

```python
#N95 Respirator No Longer Available in 7 Days
N95_mask_avail=df1.groupby(['Week Ending','Provider State','N95 Respirator No Longer Available in 7 Days']).size().reset_index().rename(columns={0:'Count of Nursing Home having N95'})
N95_mask_avail=N95_mask_avail[N95_mask_avail['N95 Respirator No Longer Available in 7 Days']=='N']
N95_mask_avail.head()
#N95_mask_avail.shape
```

```python
## Face Masks No Longer Available in 7 Days
```

```python
#Face Masks No Longer Available in 7 Days
face_mask_avail=df1.groupby(['Week Ending','Provider State','Face Masks No Longer Available in 7 Days']).size().reset_index().rename(columns={0:'Count of Nursing Home having Face Mask'})
face_mask_avail=face_mask_avail[face_mask_avail['Face Masks No Longer Available in 7 Days']=='N']
face_mask_avail.head()
```

```python
dfs=[eye_protection_avail,gloves_avail,alcohol_avail,N95_mask_avail,face_mask_avail]
dataset=reduce(lambda left,right:pd.merge(left,right,on=['Week Ending','Provider State'],how='inner'),dfs)
```

```python
dataset.head()
```

```python
## Count of Nursing Homes in Each State
```

```python
df3=df[['Federal Provider Number','Provider Name','Provider State']]
df3.head()
df4=df3.drop_duplicates()
df4.head()
```

```python
state_wise_nursing_home_count=df4['Provider State'].value_counts().reset_index().rename(columns={'index':'Provider State','Provider State':'Count_of Nursin Home'})
state_wise_nursing_home_count.head()
```

```python
state_wise_nursing_home_count.to_csv(r'C:\Users\rutuj\OneDrive\Documents\Northeastern University MS in Data Analytics Engineering\5. Computation & Visualization\Hackathon\COVID_19_Nursing_Home_Data_03_20_2022(3)\state_wise_nursing_home_count.csv')
```

```python
dataset_1=pd.merge(dataset,state_wise_nursing_home_count,on='Provider State',how='inner')
```

```python
dataset_1.head()
```

```python
dataset_1['Eye Protection Availability(%)']=(dataset_1['Count of Eye Protection']/dataset_1['Count_of Nursin Home'])*100
dataset_1['Gloves Availability(%)']=(dataset_1['Count of Nursing Homes having Gloves']/dataset_1['Count_of Nursin Home'])*100
dataset_1['Hand Sanitizer Availability(%)']=(dataset_1['Count of Nursing Homes Having Sanitizer']/dataset_1['Count_of Nursin Home'])*100
dataset_1['N95 Mask Availability(%)']=(dataset_1['Count of Nursing Home having N95']/dataset_1['Count_of Nursin Home'])*100
dataset_1['Surgical Mask Availability']=(dataset_1['Count of Nursing Home having Face Mask']/dataset_1['Count_of Nursin Home'])*100
```
