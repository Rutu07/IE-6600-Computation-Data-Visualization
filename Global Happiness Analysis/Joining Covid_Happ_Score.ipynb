```python
df_2019_2020=pd.merge(happiness_2019[['Country','Happiness score']],happiness_2020['Ladder score'],left_on=happiness_2019['Country'],right_on=happiness_2020['Country name'],how='inner')
```

```python
df_2019_2020.head()
```

```python
df_2019_2020.rename(columns={'Happiness score':'Happiness score_2019','Ladder score':'Happiness score_2020'},inplace=True)
```

```python
df_2019_2020['Difference']=df_2019_2020['Happiness score_2020']-df_2019_2020['Happiness score_2019']
```

```python
df_2019_2020['Difference %']=(((df_2019_2020['Happiness score_2020']-df_2019_2020['Happiness score_2019'])/df_2019_2020['Happiness score_2019'])*100)
```
